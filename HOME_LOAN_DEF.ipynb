```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from imblearn.combine import SMOTETomek
from imblearn.over_sampling import ADASYN
from imblearn.pipeline import Pipeline as imba_Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
import lightgbm as lgb
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import classification_report, precision_recall_curve
from tqdm import tqdm

cat_cols = [
    'NAME_CONTRACT_TYPE', 'CODE_GENDER', 'FLAG_OWN_CAR', 'FLAG_OWN_REALTY', 
    'NAME_TYPE_SUITE', 'NAME_INCOME_TYPE', 'NAME_EDUCATION_TYPE', 'NAME_FAMILY_STATUS', 
    'NAME_HOUSING_TYPE', 'OCCUPATION_TYPE', 'WEEKDAY_APPR_PROCESS_START', 'ORGANIZATION_TYPE', 
    'FONDKAPREMONT_MODE', 'HOUSETYPE_MODE', 'WALLSMATERIAL_MODE', 'EMERGENCYSTATE_MODE'
]

num_cols = application_train.select_dtypes(include=['float64', 'int64']).columns.tolist()
num_cols = [col for col in num_cols if col != 'TARGET']

def target_encode_and_add_columns(df, target_col, cat_cols):
    target_means = {}
    for col in cat_cols:
        means = df.groupby(col)[target_col].mean()
        new_col_name = f"{col}_target_encoded"
        df[new_col_name] = df[col].map(means)
        target_means[col] = means
    return df, target_means

application_train_encoded, _ = target_encode_and_add_columns(application_train, 'TARGET', cat_cols)

# Drop original categorical columns
application_train_encoded = application_train_encoded.drop(columns=cat_cols)

# Split numerical and target columns
X = application_train_encoded.drop(columns=['TARGET'])
y = application_train_encoded['TARGET']

# Normalize numerical features using MinMaxScaler
scaler = MinMaxScaler()
X[num_cols] = scaler.fit_transform(X[num_cols])

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Define models with class weights
models = {
    'Logistic Regression': (LogisticRegression(solver='liblinear', class_weight='balanced'), {
        'logistic_regression__C': [0.1, 1]
    }),
    'Random Forest': (RandomForestClassifier(class_weight='balanced'), {
        'random_forest__n_estimators': [50],
        'random_forest__max_depth': [None]
    }),
    'Gradient Boosting': (GradientBoostingClassifier(), {
        'gradient_boosting__n_estimators': [50],
        'gradient_boosting__learning_rate': [0.1]
    }),
    'XGBoost': (XGBClassifier(use_label_encoder=False, eval_metric='logloss', 
                              scale_pos_weight=len(y_train[y_train == 0]) / len(y_train[y_train == 1])), {
        'xgboost__n_estimators': [50],
        'xgboost__learning_rate': [0.1]
    }),
    'LightGBM': (lgb.LGBMClassifier(), {
        'lightgbm__n_estimators': [50],
        'lightgbm__learning_rate': [0.1]
    })
}

best_models = {}
results = {}

for name, (model, params) in tqdm(models.items(), desc="Models", unit="model"):
    print(f"Running RandomizedSearchCV for {name}...")
    
    pipeline = imba_Pipeline([
        ('adasyn', ADASYN(random_state=42)), 
        (name.lower().replace(' ', '_'), model)  
    ])
    
    # RandomizedSearchCV with fewer iterations and recall as the scoring metric
    random_search = RandomizedSearchCV(pipeline, params, n_iter=3, cv=3, scoring='recall', 
                                       verbose=1, random_state=42, n_jobs=-1, return_train_score=True)
    
    random_search.fit(X_train, y_train)
    
    # Best model and parameters
    best_models[name] = random_search.best_estimator_
    results[name] = {
        'best_params': random_search.best_params_,
        'best_score': random_search.best_score_,
        'test_score': best_models[name].score(X_test, y_test)
    }

    print(f"{name} Best Params: {results[name]['best_params']}")
    print(f"{name} Test Score: {results[name]['test_score']}")

best_model_name = max(results, key=lambda x: results[x]['test_score'])
print(f"\nBest Model: {best_model_name}")
print(classification_report(y_test, best_models[best_model_name].predict(X_test)))

y_pred_prob = best_models[best_model_name].predict_proba(X_test)[:, 1]

precision, recall, thresholds = precision_recall_curve(y_test, y_pred_prob)
custom_threshold = 0.3  # Adjust this based on precision-recall tradeoff
y_pred_custom = (y_pred_prob >= custom_threshold).astype(int)

print(f"\nClassification report with adjusted threshold (threshold = {custom_threshold}):")
print(classification_report(y_test, y_pred_custom))

# Print summary of results
for model_name, result in results.items():
    print(f"{model_name} - Best Params: {result['best_params']}, Test Score: {result['test_score']}")
```
